Средний трафик 20000 человек в день. Посчитайте сколько дней необходимо держать эксперимент при alpha = 5% и beta = 20%._



### Определим размер выборки для сравнения двух независимых средних
1. Рассчитаем ожидаемый таймспент на юзера в день: $25+25*10\%=27.5$
